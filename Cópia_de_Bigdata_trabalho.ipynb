```python
import pandas as pd

# Carrega o arquivo Excel
file_path = '/content/bigdata Estudo hibrido.xlsx'
df = pd.read_excel(file_path)

# Remove espaços extras dos nomes das colunas
df.columns = df.columns.str.strip()

# Exibe a tabela no início
print("Tabela com os dados disponíveis:")
print(df.to_string(index=False))  # Mostra os dados completos no console sem índice

# Função para buscar informações de um estado ou múltiplos estados
def buscar_informacoes_estados(estados):
    estados = [estado.upper() for estado in estados]
    dados_selecionados = df[df['Estado'].isin(estados)]

    if not dados_selecionados.empty:
        informacoes = []
        colunas_desejadas = ['Estado', '2019 (%)', '2020 (%)', '2021 (%)', '2022 (%)', '2023 (%)']

        for _, linha in dados_selecionados.iterrows():
            informacoes_formatadas = f"Estado: {linha['Estado']}\n"
            for coluna in colunas_desejadas[1:]:
                if coluna in linha.index:
                    informacoes_formatadas += f"{coluna}: {linha[coluna]}\n"
                else:
                    informacoes_formatadas += f"{coluna}: Não encontrado\n"
            informacoes.append(informacoes_formatadas)

        return informacoes, dados_selecionados
    else:
        return f"Estado(s) '{', '.join(estados)}' não encontrado(s).", None

# Função para calcular a média de crescimento dos estados selecionados
def calcular_media_estados(dados_selecionados):
    if dados_selecionados is not None:
        colunas_anos = ['2019 (%)', '2020 (%)', '2021 (%)', '2022 (%)', '2023 (%)']
        medias = dados_selecionados[colunas_anos].mean(axis=1)
        resultados_media = [
            f"Média de crescimento dos anos para o estado {estado}: {media:.2f}%"
            for estado, media in zip(dados_selecionados['Estado'], medias)
        ]
        return "\n".join(resultados_media)
    else:
        return "Não foi possível calcular a média, pois os estados não foram encontrados."

# Entrada do usuário para selecionar estados
estados_input = input("Digite os estados que deseja buscar separados por vírgula: ")
estados = [estado.strip() for estado in estados_input.split(',')]

# Busca informações dos estados
informacoes, dados_selecionados = buscar_informacoes_estados(estados)
if isinstance(informacoes, list):
    for info in informacoes:
        print(info)
else:
    print(informacoes)

# Calcula a média dos estados
media_estados = calcular_media_estados(dados_selecionados)
print(media_estados)
```
