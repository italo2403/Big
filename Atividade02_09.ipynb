```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

data = {'Idade': [18, 23, 28, 33, 38, 43, 48, 53, 58, 63],
        'Custo': [871, 1132, 1242, 1356, 1488, 1638,
                  2130, 2454, 3066, 4090]}
df = pd.DataFrame(data)

plt.scatter(df['Idade'], df['Custo'])
plt.xlabel('Idade')
plt.ylabel('Custo')
plt.title('Relação entre Idade e Custo')
plt.show()

X = df[['Idade']]  # Variável independente
y = df['Custo']    # Variável dependente

model = LinearRegression()
model.fit(X, y)

idade_nova = 45
custo_previsto = model.predict([[idade_nova]])
print(f"O custo estimado para um paciente de 45 anos é: {custo_previsto[0]:.2f}")

idade_nova = 30
custo_previsto = model.predict([[idade_nova]])
print(f"O custo estimado para um paciente de 45 anos é: {custo_previsto[0]:.2f}")

idade_nova = 44
custo_previsto = model.predict([[idade_nova]])
print(f"O custo estimado para um paciente de 45 anos é: {custo_previsto[0]:.2f}")

idade_nova = 72
custo_previsto = model.predict([[idade_nova]])
print(f"O custo estimado para um paciente de 45 anos é: {custo_previsto[0]:.2f}")

idade_nova = 102
custo_previsto = model.predict([[idade_nova]])
print(f"O custo estimado para um paciente de 45 anos é: {custo_previsto[0]:.2f}")
```
